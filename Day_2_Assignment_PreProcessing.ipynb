```python
#preProcessing
from sklearn.preprocessing import LabelEncoder
sample=data.copy()
le = LabelEncoder()
```

```python
#one_hot_encoded
one_hot_encoded = pd.get_dummies(sample['City'], prefix='City')
sample = pd.concat([sample, one_hot_encoded], axis=1)
sample = sample.loc[:, ~sample.columns.duplicated()]
sample
```

```python
#label encoding
sample['Gender_Label'] = le.fit_transform(sample['Gender'])
sample
```

```python
sample = sample.copy()
Performance_Rating_mapping = {
  1:'Excellent',
  2:'Best',
  3:'Good',
  4:'Average',
  5:'Poor'
}
sample['Performance_Rating_Category'] = sample['Performance_Rating'].map(Performance_Rating_mapping)
sample
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
sample2= ft.copy()
```

```python
numerical_columns = ['Age','Monthly_Salary']

sample2.dropna(subset=numerical_columns,inplace=True)

minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
```

```python
#minMax
sample2['Age_MinMax'] = minmax_scaler.fit_transform(sample2[['Age']])
sample2['Monthly_Salary_MinMax'] = minmax_scaler.fit_transform(sample2[['Monthly_Salary']])

sample2['Age_Standard'] = standard_scaler.fit_transform(sample2[['Age']])
sample2['Monthly_Salary_Standard'] = standard_scaler.fit_transform(sample2[['Monthly_Salary']])
```
