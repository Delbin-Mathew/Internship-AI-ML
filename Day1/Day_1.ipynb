```python
#New Column creation based on average mean
titanic['Age_mean_imputed'] = titanic['Age'].fillna(titanic['Age'].mean())
titanic['Age_median_imputed'] = titanic['Age'].fillna(titanic['Age'].median())
titanic.head(13)
```

```python
titanic['Cabin_mode'] = titanic['Cabin'].fillna(titanic['Cabin'].mode()[0])
titanic.head(10)
```

```python
#for data visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style = "whitegrid")
plt.figure(figsize= (10,6))

sns.boxplot(data = titanic, x = 'Pclass', y = 'Age',palette = 'pastel')
plt.title('Box Plot of Age by Passenger Class',fontsize = 16)
plt.xlabel('Passenger Class',fontsize = 12)
plt.ylabel('Age',fontsize = 12)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))

sns.set(style="whitegrid")

plt.figure(figsize=(10,6))

sns.barplot(data=titanic,x='Sex',y='Survived',palette='pastel')

plt.title('Survival Rate of Male and Female Passengers', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
```

```python
titanic[['Survived','Sex']]
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))

sns.lineplot(data=titanic, x='Embarked', y='Fare', palette='pastel')

plt.title('Fare by Embarked Location', fontsize=16)
plt.xlabel('Embarked', fontsize=12)
plt.ylabel('Fare', fontsize=12)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
titanic['Sex_Label'] = label_encoder.fit_transform(titanic['Sex'])

one_hot_encoded = pd.get_dummies(titanic['Embarked'], prefix='Embarked')
titanic = pd.concat([titanic, one_hot_encoded], axis=1)

# Since 'Pclass' is ordinal, we can create a mapped column for demonstration.
pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third'}
# 3. Ordinal Encoding For p class
titanic['Pclass_Category'] = titanic['Pclass'].map(pclass_mapping)

# Display the resulting dataset with relevant encoded columns
print(titanic[['Sex', 'Sex_Label', 'Embarked', 'Embarked_Q', 'Embarked_S', 'Embarked_C', 'Pclass', 'Pclass_Category']].head())
```
